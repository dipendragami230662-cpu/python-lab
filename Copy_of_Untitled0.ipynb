## 3D Plots and Heatmaps

### 3D Scatter Plot

A 3D scatter plot is used to display the relationship between three numerical variables. Each point is defined by its x, y, and z coordinates.

We'll use `matplotlib.pyplot` and its `mplot3d` toolkit for this.


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate some sample 3D data
x_3d = np.random.rand(100)
y_3d = np.random.rand(100)
z_3d = np.random.rand(100)

# Create a new figure and add a 3D subplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter points
ax.scatter(x_3d, y_3d, z_3d, c='blue', marker='o')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Simple 3D Scatter Plot')

# Show the plot
plt.show()
```

### Heatmap

A heatmap is a graphical representation of data where individual values contained in a matrix are represented as colors. It's often used to visualize correlation matrices or to show data intensity.

We'll use the `seaborn` library for creating heatmaps, which provides a high-level interface for drawing attractive statistical graphics.


```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data for a heatmap (e.g., a correlation matrix)
data = np.random.rand(10, 10)
df_heatmap = pd.DataFrame(data, columns=[f'col_{i}' for i in range(10)], index=[f'row_{i}' for i in range(10)])

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_heatmap, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Sample Heatmap')
plt.show()
```

## Exploring Various Plot Types

Here are several common plot types to help you understand and visualize different aspects of your data.

### Line Plot

A line plot is ideal for showing trends over time or ordered categories. It connects data points with straight line segments.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for a line plot (e.g., a simple trend over time)
x_line = np.linspace(0, 10, 50) # 50 points from 0 to 10
y_line = np.sin(x_line) + np.random.normal(0, 0.1, 50) # Sine wave with some noise

plt.figure(figsize=(10, 6))
plt.plot(x_line, y_line, marker='o', linestyle='-', color='green', label='Data Trend')
plt.title('Sample Line Plot (Trend Over X-axis)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

### Bar Plot

Bar plots are excellent for comparing discrete categories or showing quantities for different groups. The height of each bar represents the value.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data for a bar plot
categories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [23, 45, 56, 12]

plt.figure(figsize=(8, 6))
sns.barplot(x=categories, y=values, palette='viridis')
plt.title('Sample Bar Plot (Product Sales)')
plt.xlabel('Product')
plt.ylabel('Sales Quantity')
plt.show()
```

### Histogram

A histogram displays the distribution of a single numerical variable. It divides the data into 'bins' and shows how many data points fall into each bin.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for a histogram (e.g., scores on an exam)
data_hist = np.random.normal(75, 10, 1000) # 1000 data points, mean 75, std dev 10

plt.figure(figsize=(10, 6))
plt.hist(data_hist, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Sample Histogram (Distribution of Exam Scores)')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Box Plot

Box plots (or box-and-whisker plots) are used to show the distribution of numerical data and skewness through displaying the data quartiles (or percentiles) and averages. They are also useful for identifying outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data for a box plot (e.g., performance of different algorithms)
data_box = {
    'Algorithm A': np.random.normal(0, 1, 100),
    'Algorithm B': np.random.normal(1, 1.5, 100),
    'Algorithm C': np.random.normal(-0.5, 0.8, 100)
}
df_box = pd.DataFrame(data_box)

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_box, palette='pastel')
plt.title('Sample Box Plot (Performance Across Algorithms)')
plt.xlabel('Algorithm')
plt.ylabel('Performance Metric')
plt.show()
```

### Scatter Plot (Advanced Example)

An enhanced scatter plot can show the relationship between two numerical variables, with additional dimensions represented by markers size and color. This can reveal clusters or patterns.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for a scatter plot
x_scatter_adv = np.random.rand(100) * 10
y_scatter_adv = 2 * x_scatter_adv + np.random.normal(0, 2, 100)
sizes = np.random.rand(100) * 200 + 50 # Varying marker sizes
colors = np.random.rand(100) # Varying marker colors

plt.figure(figsize=(10, 7))
scatter = plt.scatter(x_scatter_adv, y_scatter_adv, s=sizes, c=colors, cmap='coolwarm', alpha=0.7, edgecolors='w')
plt.title('Sample Scatter Plot (Relationship with Size and Color)')
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
cbar = plt.colorbar(scatter)
cbar.set_label('Additional Metric (Color Scale)')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

## Subplots

Subplots allow you to display multiple plots within the same figure, which is useful for comparing different visualizations or showing various aspects of your data side-by-side.

We use `plt.subplots()` to create a figure and a grid of subplots. It returns a figure object and an array of axes objects, one for each subplot.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Sample data for subplots
x_data = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x_data)
y_cos = np.cos(x_data)

bar_categories = ['A', 'B', 'C', 'D']
bar_values = [15, 25, 30, 20]

hist_data = np.random.randn(1000)

# Create a figure with a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Various Subplots in One Figure', fontsize=16)

# Plot 1: Line Plot (top-left)
axes[0, 0].plot(x_data, y_sin, color='purple', label='sin(x)')
axes[0, 0].plot(x_data, y_cos, color='orange', linestyle='--', label='cos(x)')
axes[0, 0].set_title('Sine and Cosine Waves')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Plot 2: Bar Plot (top-right)
sns.barplot(x=bar_categories, y=bar_values, ax=axes[0, 1], palette='pastel')
axes[0, 1].set_title('Category Distribution')
axes[0, 1].set_xlabel('Category')
axes[0, 1].set_ylabel('Count')

# Plot 3: Histogram (bottom-left)
axes[1, 0].hist(hist_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axes[1, 0].set_title('Distribution of Random Data')
axes[1, 0].set_xlabel('Value')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].grid(axis='y', alpha=0.75)

# Plot 4: Scatter Plot (bottom-right)
scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)
scatter_sizes = np.random.rand(50) * 100 + 20
scatter_colors = np.random.rand(50)
axes[1, 1].scatter(scatter_x, scatter_y, s=scatter_sizes, c=scatter_colors, cmap='viridis', alpha=0.8)
axes[1, 1].set_title('Random Scatter Plot')
axes[1, 1].set_xlabel('Feature A')
axes[1, 1].set_ylabel('Feature B')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle
plt.show()
```

## Bubble Chart

A bubble chart is a variation of a scatter plot where, in addition to the x and y coordinates, a third variable is represented by the size of the 'bubble' or marker. A fourth variable can also be represented by the color of the bubbles.

This type of chart is useful for visualizing the relationship between three or four quantitative variables, making it easy to spot clusters or trends in multi-dimensional data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for a bubble chart
num_points = 50
x_bubble = np.random.rand(num_points) * 10 # First variable (x-axis)
y_bubble = np.random.rand(num_points) * 10 # Second variable (y-axis)
sizes_bubble = np.random.rand(num_points) * 1000 + 100 # Third variable (size of bubble)
colors_bubble = np.random.rand(num_points) # Fourth variable (color of bubble)

plt.figure(figsize=(12, 8))
scatter = plt.scatter(x_bubble, y_bubble, s=sizes_bubble, c=colors_bubble, cmap='viridis', alpha=0.7, edgecolors='w', linewidth=0.5)

# Add labels and title
plt.xlabel('Variable 1 (X-axis)')
plt.ylabel('Variable 2 (Y-axis)')
plt.title('Sample Bubble Chart (Variable 3 represented by size, Variable 4 by color)')

# Add a color bar to explain the color mapping
cbar = plt.colorbar(scatter)
cbar.set_label('Variable 4 (Color Scale)')

plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
