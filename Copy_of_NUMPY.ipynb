## 6. Arithmetic Operations

NumPy arrays support element-wise arithmetic operations, making it easy to perform calculations on entire arrays without explicit loops.

```
arr = np.arange(0,11)
print("Array:",arr)
print("Element at 7th index is:", arr[7])
print("Elements from 3rd to 5th index are:", arr[3:6]) # Slicing from index 3 up to (but not including) 6
print("Elements up to 4th index are:", arr[:4]) # Slicing from the beginning up to (but not including) 4
print("Elements from last backwards are:", arr[-1::-1]) # Reverse the array
print("3 Elements from last backwards are:", arr[-1:-6:-2]) # Every second element from the end, starting at the last

arr = np.arange(0,21,2)
print("New array:",arr)
print("Elements at 2nd, 4th, and 9th index are:", arr[[2,4,9]]) # Fancy indexing: selecting elements at specific indices

mat = np.array(ri(10,100,15)).reshape(3,5) # Create a 3x5 matrix of random 2-digit numbers
print("Matrix of random 2-digit numbers\n--------------------------------\n",mat)

print("\nDouble bracket indexing\n------------------------")
print("Element in row index 1 and column index 2:", mat[1][2]) # Traditional Python list-of-lists indexing

print("\nSingle bracket with comma indexing\n----------------------------------")
print("Element in row index 1 and column index 2:", mat[1,2]) # NumPy's preferred and more efficient indexing
print("\nRow or column extract\n----------------------")

print("Entire row at index 2:", mat[2]) # Extracting a full row
print("Entire column at index 3:", mat[:,3]) # Extracting a full column (all rows, column index 3)

print("\nSubsetting sub-matrices\n--------------------------")
print("Matrix with row indices 1 and 2 and column indices 3 and 4\n", mat[1:3,3:5]) # Slicing rows and columns to get a sub-matrix
print("Matrix with row indices 0 and 1 and column indices 1 and 3\n", mat[0:2,[1,3]]) # Combining slicing for rows and fancy indexing for columns
```

## 7. Broadcasting

Broadcasting is a powerful feature in NumPy that allows arithmetic operations to be performed on arrays of different shapes. It automatically 'stretches' smaller arrays to match the shape of larger arrays during operations, provided certain compatibility rules are met.

For more detailed information, refer to the NumPy documentation on broadcasting: [https://docs.scipy.org/doc/numpy-1.10.1/user/basics.broadcasting.html](https://docs.scipy.org/doc/numpy-1.10.1/user/basics.broadcasting.html)

```
mat1 = np.array(ri(1,10,9)).reshape(3,3)
mat2 = np.array(ri(1,10,9)).reshape(3,3)
print("\n1st Matrix of random single-digit numbers\n----------------------------------------\n",mat1)
print("\n2nd Matrix of random single-digit numbers\n----------------------------------------\n",mat2)

print("\nAddition\n------------------\n", mat1+mat2) # Element-wise addition
print("\nMultiplication\n------------------\n", mat1*mat2) # Element-wise multiplication
print("\nDivision\n------------------\n", mat1/mat2) # Element-wise division
print("\nLinear combination: 3*A - 2*B\n-----------------------------\n", 3*mat1-2*mat2) # Scalar multiplication and subtraction

print("\nAddition of a scalar (100)\n-------------------------\n", 100+mat1) # Scalar addition (broadcasts to all elements)

print("\nExponentiation, matrix cubed here\n----------------------------------------\n", mat1**3) # Element-wise exponentiation
print("\nExponentiation, sq-root using pow function\n-------------------------------------------\n",pow(mat1,0.5)) # Element-wise square root
```

**Broadcasting (super cool feature)**
NumPy operations are usually done on pairs of arrays on an element-by-element basis. In the simplest case, the two arrays must have exactly the same shape.

NumPy’s broadcasting rule relaxes this constraint when the arrays’ shapes meet certain constraints. When operating on two arrays, NumPy compares their shapes element-wise. It starts with the trailing dimensions, and works its way forward. Two dimensions are compatible when

they are equal, or
one of them is 1
If these conditions are not met, a ValueError: frames are not aligned exception is thrown, indicating that the arrays have incompatible shapes. The size of the resulting array is the maximum size along each dimension of the input arrays.

For more detail, please look up: https://docs.scipy.org/doc/numpy-1.10.1/user/basics.broadcasting.html

```
start = np.zeros((4,3))
print(start)
```

```
# create a rank 1 ndarray with 3 values
add_rows = np.array([1, 0, 2])
print(add_rows)
```

```
y = start + add_rows  # add to each row of 'start' using broadcasting
print(y)
```

```
# create an ndarray which is 4 x 1 to broadcast across columns
add_cols = np.array([[0,1,2,3]])
add_cols = add_cols.T
print(add_cols)
```

```
# add to each column of 'start' using broadcasting
y = start + add_cols
print(y)
```

```
# this will just broadcast in both dimensions
add_scalar = np.array([100])
print(start+add_scalar)
```

## 8. Linear Algebra Operations

NumPy is a powerful tool for linear algebra, providing functions for dot products, inner/outer products, transposing matrices, calculating the trace, eigenvalues, and solving linear equations.

```
start = np.zeros((4,3)) # A 4x3 matrix of zeros
print(start)

# create a rank 1 ndarray with 3 values
add_rows = np.array([1, 0, 2])
print(add_rows)

y = start + add_rows  # add to each row of 'start' using broadcasting
print(y)

# create an ndarray which is 4 x 1 to broadcast across columns
add_cols = np.array([[0,1,2,3]])
add_cols = add_cols.T # Transpose to make it a column vector
print(add_cols)

# add to each column of 'start' using broadcasting
y = start + add_cols
print(y)

# this will just broadcast in both dimensions
add_scalar = np.array([100])
print(start+add_scalar) # Adding a scalar broadcasts it to every element
```

### 8.2 Matrix Trace

The trace of a square matrix is the sum of the elements on its main diagonal. NumPy's `np.trace` function can also compute the sum along other diagonals using an offset.

```
A = ri(1,10,9).reshape(3,3) # Create a 3x3 matrix
print("\n3x3 Matrix of random single-digit numbers\n","-"*50,"\n",A)
print("\nMatrix transpose\n","-"*50,"\n",np.transpose(A)) # Transpose of matrix A

B = ri(1,10,6).reshape(3,2) # Create a 3x2 matrix
print("\n3x2 Matrix of random single-digit numbers\n","-"*50,"\n",B)
print("\n2x3 Matrix transpose\n","-"*50,"\n",np.transpose(B)) # Transpose of matrix B
print("\nMatrix multiplication of B and B-transpose\n","-"*50,"\n",np.dot(B, np.transpose(B))) # Example of multiplying a matrix by its transpose
```

```
A = ri(1,10,16).reshape(4,4)
print("\n4x4 Matrix of random single-digit numbers\n","-"*50,"\n",A)
print("\nMatrix trace\n","-"*50,"\n",np.trace(A))
print("\nMatrix trace with ofset +1 (upper triangle)\n","-"*50,"\n",np.trace(A,offset=1))
print("\nMatrix trace with ofset -1 (lower triangle)\n","-"*50,"\n",np.trace(A,offset=-1))
```

### 8.3 Eigenvalues and Eigenvectors

Eigenvalues and eigenvectors are crucial in many areas of linear algebra, physics, and engineering. They describe the directions along which a linear transformation acts by scaling only.

```
A = ri(1,10,16).reshape(4,4) # Create a 4x4 matrix
print("\n4x4 Matrix of random single-digit numbers\n","-"*50,"\n",A)
print("\nMatrix trace\n","-"*50,"\n",np.trace(A)) # Sum of elements on the main diagonal (offset=0)
print("\nMatrix trace with offset +1 (upper triangle)\n","-"*50,"\n",np.trace(A,offset=1)) # Sum of elements on the diagonal above the main diagonal
print("\nMatrix trace with offset -1 (lower triangle)\n","-"*50,"\n",np.trace(A,offset=-1)) # Sum of elements on the diagonal below the main diagonal
```

# **Eigenvalues and eigenvectors**

```
A = np.random.randn(9).reshape(3,3)
print("Original matrix\n",A)
print("\n")
w,v=np.linalg.eig(A)
print("Eigenvalues:\n",w)
print("\n")
print("Eigenvectors:\n",v)
```

### 8.4 Linear Equation Solving and Matrix Inverse

NumPy's `np.linalg.solve` function allows direct solution of systems of linear equations, and `np.linalg.inv` can compute the inverse of a matrix.

```
A = np.random.randn(9).reshape(3,3) # Create a 3x3 matrix with random numbers from a standard normal distribution
print("Original matrix\n",A)
print("\n")
w,v=np.linalg.eig(A) # Compute eigenvalues (w) and eigenvectors (v)
print("Eigenvalues:\n",w)
print("\n")
print("Eigenvectors:\n",v)
```

# Linear equation solving, matrix inverse, linear least suqare
You can directly solve linear system of equations using NumPy writing only one line of code.

Suppose you want to solve the following set of equations for (x,y,z). It is easy to see that (2,1,5) are the unique solutions.

2x + 5y + z = 14;

3x - 2y - z = -1;

x - 3y + z = 4

```
A = np.array([[2,5,1],[3,-2,-1],[1,-3,1]])
B = np.array([14,-1,4])
x = np.linalg.solve(A,B)

print("The solutions are:",x)
```
