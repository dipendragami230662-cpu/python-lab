Question 2: Clean data and update the CSV file
Replace all column values which contain ‘?’ and n.a with NaN.

### Objective
This section focuses on data cleaning by handling missing values. The dataset contains '?' and 'n.a' as placeholders for missing data, which are not recognized by pandas as NaN (Not a Number). This step will replace these specific strings with `NaN` to allow for proper missing data handling in subsequent analysis. The cleaned data will then be saved back to the CSV file.

```python
df = pd.read_csv("Automobile_data.csv", na_values={
'price':["?","n.a"],
'stroke':["?","n.a"],
'horsepower':["?","n.a"],
'peak-rpm':["?","n.a"],
'average-mileage':["?","n.a"]})
print (df)
df.to_csv("Automobile_data.csv")
```

```python
#Question 6: Find each company’s Higesht price car
```

### Objective
This section finds the highest-priced car for each manufacturing company. It involves grouping the DataFrame by 'company' and then applying the `max()` aggregation function to the 'price' column for each group.

```python
import pandas as pd
df = pd.read_csv("Automobile_data.csv")
car_Manufacturers = df.groupby('company')
priceDf = car_Manufacturers[['company','price']].max()
priceDf
```

```python
#Question 9: Concatenate two data frames using following conditions

   # Create two data frames using following two Dicts, Concatenate those two data frame and create a key for each data frame.

#GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
```

### Objective
This section demonstrates how to concatenate two separate DataFrames. Two dictionaries, `GermanCars` and `japaneseCars`, are used to create two DataFrames. These are then combined into a single DataFrame using `pd.concat()`, with a `keys` argument to create a multi-index that distinguishes between 'German' and 'Japanese' cars.

```python
import pandas as pd
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
carsDf1 = pd.DataFrame.from_dict(GermanCars)
```

```python
# Create DataFrame for Japanese Cars
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
carsDf2 = pd.DataFrame.from_dict(japaneseCars)

# Concatenate the two DataFrames with keys
# The 'keys' argument adds a hierarchical index (multi-index) to the resulting DataFrame,
# making it easy to identify which original DataFrame each row came from.
cars = pd.concat([carsDf1, carsDf2], keys=["German", "Japanese"])
display(cars)
```

```python
#Question 10: Merge two data frames using following condition
   # Create two data frames using following two Dicts, Merge two data frames, and append second data frame as a new column to first data frame.

#Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
#car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

### Objective
This section illustrates how to merge two DataFrames based on a common column. Two dictionaries, `Car_Price` and `car_Horsepower`, are used to create two DataFrames. These DataFrames are then merged using `pd.merge()` on the 'Company' column, effectively appending the 'horsepower' information as a new column to the car price data.

```python
import pandas as pd
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
carPriceDf = pd.DataFrame.from_dict(Car_Price)
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
carsHorsepowerDf = pd.DataFrame.from_dict(car_Horsepower)
carsDf = pd.merge(carPriceDf, carsHorsepowerDf, on="Company")
carsDf
```
